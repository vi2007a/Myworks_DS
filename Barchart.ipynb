```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.bar(x = df_used_cars["Brand"],
           height = df_used_cars["Cars Listings"])
plt.show()
```

```python
plt.figure(figsize = (9 , 6))
plt.bar(x = df_used_cars["Brand"],
           height = df_used_cars["Cars Listings"])
plt.show()
```

```python
plt.figure(figsize = (9 , 6))
plt.bar(x = df_used_cars["Brand"],
           height = df_used_cars["Cars Listings"])
plt.xticks(rotation = 45)
plt.show()
```
